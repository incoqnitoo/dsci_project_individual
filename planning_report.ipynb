<font size="5"> Exploratory Visualizations

```r
options(repr.plot.width = 6, repr.plot.height = 8)
age_sub_plot <- players |>
    filter(played_hours > 0) |> # avoiding the introduction of infinite values as log(0) DNE
    ggplot(aes(x = Age, y = played_hours)) +
    geom_point() + 
    labs(x = "Age (in years)", y = "Playtime (in hours)") +
    ggtitle("Age vs Playtime") +
    theme(text = element_text(size = 16)) +
    scale_y_log10()

age_sub_plot
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
age_dist <- players |>
    ggplot(aes(x = Age)) +
    geom_histogram() +
    labs(x = "Player age (in years)", y = "Number of players counted within the dataset") +
    ggtitle("Distribution of player range in the 'players' dataset") +
    theme(text = element_text(size = 20))

age_dist
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
playtime_dist <- players |> 
    filter(played_hours > 0) |>
    ggplot(aes(x = played_hours)) +
    geom_histogram(binwidth = 10) +
    labs(x = "Playtime (hours)", y = "Number of the corresponding playtime") +
    scale_y_log10() +
    ggtitle("Distribution of playtime in the 'players' dataset") +
    theme(text = element_text(size = 20))
    
playtime_dist
```

<font size="5"> Exploratory Plot Observations

- There is no clear relationship between age and playtime.
- The age distribution is concentrated at ~15-25 years, this range provides most information for the training data.
- Most players play for ~0-30 hours, this range provides the most information for the training data.

# **Methods and Plan**

<font size="5"> K-NN Classification

- K-NN Classification will be used to address this question since a categorical variable is predicted based on two numerical predictor variables. 
- Data split: 70% training, 30% testing, stratified by "subscribe".
- Data pre-processing will be done by standardization.
- Tuning the model by 5-fold cross-validation on the training data to find the K that minimizes error. 
- The model will be evaluated on the testing set and evaluated using a confusion matrix. 

<font size="5"> Potential Limitations

- The predictor variables - "played_hours" and "Age" -  may have too little influence on subscription status. Prediction model might be ineffective.

- ~70% of players are subscribed (TRUE). This may lead to a bias within the model to perdict subscribed with a bias, which may lead to a low recall if not subscribed (FALSE) is the interested variable.
